```python
import pandas as pd
import numpy as np

#크롤링 후 csv파일로 저장 -> 파일불러오기 후 전처리

df = pd.read_csv("C:/Users/sejung/Downloads/auction_test5.csv",encoding="utf-8")
print(df)
```

```python
df1= df.copy()
df1["유찰수"] = df1["유찰횟수"].str.split(" ").str.get(0) # split띄어쓰기를 기준으로 get 앞0 뒤1 을 가져와라
print(df1)
```

```python
df3 = df2.drop("유찰횟수", axis = 1)
print(df3)
```
