```python
import requests
import re #정규식
from bs4 import BeautifulSoup

url = "https://www.shinhanauctionsa.com/auction/ca_list.php"
res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text,"lxml")

trs =  soup.find_all("tr", attrs={"class":"HighlightableRow"})
for tr in trs:
    tds = tr.find_all("td")
    if len(tds) >= 5:
        number = tds[1].find_all("span")[0].get_text().strip() #경매번호
        types = tds[1].find_all("span")[1].get_text().strip() #매물종류
        price =  tds[2].find_all("dd")[0].get_text().strip() #감정가
        lowest = tds[2].find_all("dd")[1].get_text().strip() #최저가
        fail = tds[3].find_all("dd")[1].get_text().strip().replace("\t", "") #유찰횟수 & 감정가대비 유찰가 # 불필요 과한 공백 제거
        views = tds[5].find_all("div")[1].get_text().strip() #조회수

        print(number, types, price, lowest, fail, views)
    else:
        print('데이터가 다름!!!')
```
