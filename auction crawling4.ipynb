```python
import requests
from bs4 import BeautifulSoup

url = "https://www.shinhanauctionsa.com/auction/ca_list.php?page_code=9012"
res = requests.get(url)
res.raise_for_status()

#soup = BeautifulSoup(response.text,'html.parser')
soup = BeautifulSoup(res.text,"lxml") #html 문서를 lxml 파서를 통해서 뷰숩객체를 가지고 온 것
#print(soup.title)
#print(soup.title.get_text()) #soup은 페이지의 모든 정보를 가지고 있음
#print(soup.a) #처음 발견된 a엘리먼트에 대한 정보를 모두 뿌려줘[<a href]
#print(soup.a.attrs) #a엘리먼트의 속성정보 반환 딕셔너리형태로 출력됨
print(soup.a["href"]) # a엘리먼트의 href 속성 값 출력
```
