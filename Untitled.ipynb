```python
import requests
import re #정규식 
from bs4 import BeautifulSoup
import csv
import pandas as pd

#CSV파일로 저장하지 않으면서 웹크롤링한 데이터 가공하기

#header = {"User-Agent":""User Agent strin검색"} 서버가 거부할 경우 사람이 직접 접속하는 것 처럼 변경


url = "https://www.shinhanauctionsa.com/auction/ca_list.php"
res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text,"lxml")

items =  soup.find_all("tr", attrs={"class":"HighlightableRow"})
for item in items:
    number = item.find("span", attrs={"class":"no f14 bold list_class"}).get_text().strip() #경매번호
    types = item.find("span", attrs={"class":"list_class f12 center white"}).get_text().strip() #매물종류
    #매물종류 제외
    #if "아파트" in types:
        #print("<아파트 매물 제외합니다")
        #continue
    price =  item.find("dd", attrs={"class":"bold"}).get_text().strip() #감정가
    lowest = item.find("dd", attrs={"class":"bold clr_blue"}).get_text().strip() #최저가
     #유찰횟수 replace("신건", "신건0회") 특정글자교체
    fail = item.find("dd", attrs={"style":";padding-bottom:5px"}).get_text().strip().replace("\t", "").replace("신건", "신건0회") .replace("(", " ").replace(")", "").strip("유찰").strip("신건")
    #조회수 strip("조회").strip(": ")특정글자삭제
    views = item.find("div", attrs={"style":"padding-top:5px"}).get_text().strip().strip("조회").strip(": ")
    
    data = (number, types, price, lowest, fail, views)
    print(data)
```

```python
data2 = pd.DataFrame(data)
print(data2)
```
