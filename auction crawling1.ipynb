```python
import requests
from bs4 import BeautifulSoup

url = "https://www.shinhanauctionsa.com/auction/ca_list.php"
res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text,"lxml")
print(soup.title)
print(soup.title.get_text())
```

```python
print(soup.find("span", attrs={"class":"no f14 bold list_class"}))
```

```python
print(soup.find("div",attres={"class":"list_addr"}))
```

```python
auctions = soup.find_all("div",attres={"class":"list_addr"})
for auction in auctions:
    print(auction.get_text())
```

```python
auctions = soup.find_all("div",attres={"class":"list_addr"})
list_addr = auctions[1].span.get_text()
print(list_addr)
```

```python
auctions = soup.find_all("div",attres={"class":"list_addr"})

for auction in auctions:
    list_addr = auction.span.get_text()
    print(list_addr)
```

```python
auctions = soup.find_all("div",attres={"class":"list_addr"})

for auction in auctions:
    rate = auction.find("span").get_text()
    print(rate)
```
