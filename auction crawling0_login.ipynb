```python
#라이브러리 import

import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
#세션만들기
session=requests.session()
```

```python
#로그인 하는 페이지의 general-requestURL에서 url 가져옴
url="https://www.shinhanauctionsa.com/member/cert.php"
```

```python
#로그인 후 가져오고 싶은 데이터 (form data)
data={
    "return_url":"https://www.shinhanauctionsa.com/index.php",
    "client_id":"alskdj7",
    "passwd":"wjfeodlfhs7@"
}
response=session.post(url,data=data) 
```

```python
#로그인 실행
response.raise_for_status()
```

```python
#크롤링할 페이지 접근
url="https://www.shinhanauctionsa.com/auction/ca_list.php?page_code=9012"
response=session.get(url)
response.raise_for_status()
#print(response.text)
```

```python
#HTML분석
soup=BeautifulSoup(response.text,'html.parser')
```

```python
#크롤링할 구간 지정
number = soup.select('no f14 bold list_class')
auc_number = [number.text for number in number]
address = soup.select('addr selectorgadget_selected')
auc_address = [address.text for address in address]
size = soup.select('addr_point blue selectorgadget_selected')
auc_size = [size.text for size in size]
```

```python
#리스트로 출력
import pandas as pd
auc = pd.DataFrame({
    "number": auc_number,
    "address": auc_address,
    "size": auc_size,
})
auc
```

```python
#크롤링할 구간 지정

number = soup.select('.no f14 bold list_class selectorgadget_selected')
address = soup.select('.addr selectorgadget_selected')
size = soup.select('.addr_point blue')


#리스트로 저장
auc_number =[]
auc_address =[]
auc_size =[]

for dom in number:
    auc_number.append(dom.text)
for dom in address:
    auc_address.append(dom.text)
for dom in size:
    auc_size.append(dom.text)       
commentLength = len(auc_number)

for i in range(commentLength):
    print(auc_number[i]+auc_address[i]+auc_size[i])
```
